```python
import pandas as pd
```

```python
import glob
```

```python
list_of_files = glob.glob("*_bcell_fr*")
```

```python
dfs = []
for file in list_of_files:
    df = pd.read_table(file, index_col=0)
    dfs.append(df)
```

```python
df = pd.concat(dfs)
```

```python
df.to_csv('TBD_b_cell_fractions.tab', sep='\t')
```
